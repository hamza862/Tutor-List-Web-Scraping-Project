```python
## Importing the required libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
## Let's load the webpage

headers = {
    'Accept-Encoding': 'gzip, deflate, sdch',
    'Accept-Language': 'en-US,en;q=0.8',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
}


r = requests.get('https://www.bdadyslexia.org.uk/dyslexia/tutor-list/search', headers=headers)
```

```python
## Create a beautifu Soup object

soup = bs(r.content)

## Let's Print out the webpage
print(soup.prettify())
```

```python
## Let's select all the listing rows
listing_rows = soup.find_all('div', class_='listing-row')
listing_rows
```

```python
## Let's extract all the names from the listing rows

Name=[]
for index, row in enumerate(listing_rows):
    name_text = listing_rows[index].find('h2').get_text()
    Name.append(name_text)
    
## Let's print out all the names
Name
```

```python
## Let's check the length of the Name
len(Name)
```

```python
## Let's extract all the locations from the listing rows
location_tags = soup.find_all('span', class_='contact-info__icon contact-info__icon--location') 

Location = []
for index, location in enumerate(location_tags):
    location_text = location_tags[index].get_text()
    Location.append(location_text)
    
## Let's print out all the locations
Location
```

```python
## Let's check the length of the location
len(Location)
```

```python
## Let's extract all the emails from the listing rows
email_tags = soup.find_all('span', class_='contact-info__icon contact-info__icon--email')

Email=[]
for index, email in enumerate(email_tags):
    email_text = email_tags[index].get_text()
    Email.append(email_text)

## Let's print out all the emails
Email
```

```python
## Let's check the length of the email
len(Email)
```

```python
## Let's extract all the phone numbers from the listing rows
contact_info = soup.find_all('div', class_='contact-info')

Phone = []
for index, row in enumerate(contact_info):
    if len(contact_info[index].find_all('span')) == 3:
        phone_text = contact_info[index].find('span', class_='contact-info__icon contact-info__icon--phone').get_text()
        Phone.append(phone_text)
    else:
        Phone.append('N/A')
        
## Let's print out all the phone numbers
Phone
```

```python
## Checking the length of the Phone
len(Phone)
```

```python
## Now let's create and store all the data in dataframe
data = {
    'Name' : Name,
    'Location' : Location,
    'Email' : Email,
    'Phone' : Phone
}

df = pd.DataFrame(data)
df.head()
```

```python
## Let's create a csv file of the data
df.to_csv('data.csv')
```
